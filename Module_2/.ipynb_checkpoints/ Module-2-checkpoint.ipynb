```python
import numpy as np
import matplotlib.pyplot as plt
```

# Activity - 2
## Analysis - 1.

```python
cov = np.array([[4, 2], [2, 16]])
mean = np.array([0, 0])
```

```python
events = np.random.multivariate_normal(mean, cov, 1000)
x_coord, y_coord = events[:,0], events[:,1]
```

```python
#print(x_coord, y_coord)
```

## Analysis - 2.

```python
plt.scatter(x_coord, y_coord, c='r')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
